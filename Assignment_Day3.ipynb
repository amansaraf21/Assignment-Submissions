### Relplot

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal',hue='event', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal',hue='event',size='region', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal',hue='region', size='event', kind='line', data=fmri, ci='sd')
```

```python
sns.relplot(x='timepoint', y='signal',hue='event', size='event', col='region', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal',hue='event', size='event', col='region',height=4, aspect=.7, kind='line', data=fmri)
```

```python
q = sns.relplot(x='timepoint', y='signal',hue='event', size='event', col='region',height=4, aspect=.7, kind='line', data=fmri)
```

### BoxPlot

```python
sns.catplot(x='region',y='signal',kind='box',data=fmri,hue='event')
```

```python
sns.catplot(x='region',y='signal',kind='box',data=fmri,hue='event',dodge=False)
```

```python
sns.catplot(x='event',y='timepoint',kind='box',data=fmri,hue='region')
```

```python
sns.catplot(x='region',y='signal',kind='boxen',data=fmri,hue='event',dodge=False)
```

### LinePLot

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',data=fmri)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='region',data=fmri,style='event')
```

```python
z = fmri.query("region == 'frontal'")
```

```python
sns.lineplot(
    data=fmri.query("region == 'frontal'"),
    x="timepoint", y="signal", hue="event", units="subject",
    estimator=None, lw=1,
)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='region',data=z,style='event')
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", err_style="bars", ci=68)
```
